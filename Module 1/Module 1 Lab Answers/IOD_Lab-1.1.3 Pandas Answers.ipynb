# 4. Visualisation


#### 4.A Plot a histogram of `age`

````python
plt.hist(df.age, bins=25, color='b', alpha=0.5,)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency (count)")
plt.show()
````

````python
num_bins = 50
n, bins, patches = plt.hist(df.age, num_bins, facecolor='blue', alpha=0.5)
````

#### 4.B BoxPlot of `age`

````python
df_plot = df.groupby('city_name', as_index=False).age.median()
````

````python
df_plot
````

````python
df_plot.sort_values('age', inplace=True)
````

````python
plt.figure(figsize=(10,10))
plt.barh(df_plot.city_name, df_plot.age)
````
